### Research Question 2: How many books were published in print vs. online resources?

```python
#Comparing digital books to physical books buy adding how many times relevant keywords appear in column

def books_format():
  physical_books = df['Physical description'].str.contains('volume').sum() + df['Physical description'].str.contains('pages').sum()
  digital_books = df['Physical description'].str.contains('online').sum()

  print(physical_books, "are physical books and" , digital_books, "are digital books.")
books_format()
```

Do any of these books contain Illustrations as well?

```python
def books_illustrated():
  illustrations = df['Physical description'].str.contains('illustrations').sum()
  if illustrations > 1:
    return True, illustrations 
books_illustrated()
```

```python
#plotting bar chart

import matplotlib.pyplot as plt
   
country = ['Physical Books', 'Digtal Books','Boooks with illustrations']
gdp_per_capita = [780, 91,356]

colors = ['purple','teal', 'red']
plt.bar(country, gdp_per_capita, color=colors)
plt.title('Physical Book Copies Vs Digital Book Copies')
plt.xlabel('Book Format')
plt.ylabel('Copies Available')
plt.show()
```
